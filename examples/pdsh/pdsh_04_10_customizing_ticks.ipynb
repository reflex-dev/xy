# Customizing Ticks

## Major and Minor Ticks

```python
# Matplotlib reference
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")
import numpy as np
```

```python
# xy.pyplot
import xy.pyplot as xy_plt

xy_plt.style.use("default")
import numpy as np
```

```python
# Matplotlib reference
mpl_ax = mpl_plt.axes(xscale="log", yscale="log")
mpl_ax.set(xlim=(1, 1e3), ylim=(1, 1e3))
mpl_ax.grid(True);
```

```python
# xy.pyplot
xy_ax = xy_plt.axes(xscale="log", yscale="log")
xy_ax.set(xlim=(1, 1e3), ylim=(1, 1e3))
xy_ax.grid(True);
```

```python
# Matplotlib reference
print(mpl_ax.xaxis.get_major_locator())
print(mpl_ax.xaxis.get_minor_locator())
```

```python
# xy.pyplot
print(xy_ax.xaxis.get_major_locator())
print(xy_ax.xaxis.get_minor_locator())
```

```python
# Matplotlib reference
print(mpl_ax.xaxis.get_major_formatter())
print(mpl_ax.xaxis.get_minor_formatter())
```

```python
# xy.pyplot
print(xy_ax.xaxis.get_major_formatter())
print(xy_ax.xaxis.get_minor_formatter())
```

## Hiding Ticks or Labels

```python
# Matplotlib reference
mpl_ax = mpl_plt.axes()
rng = np.random.default_rng(1701)
mpl_ax.plot(rng.random(50))
mpl_ax.grid()

mpl_ax.yaxis.set_major_locator(mpl_plt.NullLocator())
mpl_ax.xaxis.set_major_formatter(mpl_plt.NullFormatter())
```

```python
# xy.pyplot
xy_ax = xy_plt.axes()
rng = np.random.default_rng(1701)
xy_ax.plot(rng.random(50))
xy_ax.grid()

xy_ax.yaxis.set_major_locator(xy_plt.NullLocator())
xy_ax.xaxis.set_major_formatter(xy_plt.NullFormatter())
```

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots(5, 5, figsize=(5, 5))
mpl_fig.subplots_adjust(hspace=0, wspace=0)

# Get some face data from Scikit-Learn
from sklearn.datasets import fetch_olivetti_faces

faces = fetch_olivetti_faces().images

for i in range(5):
    for j in range(5):
        mpl_ax[i, j].xaxis.set_major_locator(mpl_plt.NullLocator())
        mpl_ax[i, j].yaxis.set_major_locator(mpl_plt.NullLocator())
        mpl_ax[i, j].imshow(faces[10 * i + j], cmap="binary_r")
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots(5, 5, figsize=(5, 5))
xy_fig.subplots_adjust(hspace=0, wspace=0)

# Get some face data from Scikit-Learn
from sklearn.datasets import fetch_olivetti_faces

faces = fetch_olivetti_faces().images

for i in range(5):
    for j in range(5):
        xy_ax[i, j].xaxis.set_major_locator(xy_plt.NullLocator())
        xy_ax[i, j].yaxis.set_major_locator(xy_plt.NullLocator())
        xy_ax[i, j].imshow(faces[10 * i + j], cmap="binary_r")
```

## Reducing or Increasing the Number of Ticks

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots(4, 4, sharex=True, sharey=True)
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots(4, 4, sharex=True, sharey=True)
```

```python
# Matplotlib reference
# For every axis, set the x and y major locator
for axi in mpl_ax.flat:
    axi.xaxis.set_major_locator(mpl_plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(mpl_plt.MaxNLocator(3))
mpl_fig
```

```python
# xy.pyplot
# For every axis, set the x and y major locator
for axi in xy_ax.flat:
    axi.xaxis.set_major_locator(xy_plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(xy_plt.MaxNLocator(3))
xy_fig
```

## Fancy Tick Formats

```python
# Matplotlib reference
# Plot a sine and cosine curve
mpl_fig, mpl_ax = mpl_plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
mpl_ax.plot(x, np.sin(x), lw=3, label="Sine")
mpl_ax.plot(x, np.cos(x), lw=3, label="Cosine")

# Set up mpl_grid, legend, and limits
mpl_ax.grid(True)
mpl_ax.legend(frameon=False)
mpl_ax.axis("equal")
mpl_ax.set_xlim(0, 3 * np.pi);
```

```python
# xy.pyplot
# Plot a sine and cosine curve
xy_fig, xy_ax = xy_plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
xy_ax.plot(x, np.sin(x), lw=3, label="Sine")
xy_ax.plot(x, np.cos(x), lw=3, label="Cosine")

# Set up xy_grid, legend, and limits
xy_ax.grid(True)
xy_ax.legend(frameon=False)
xy_ax.axis("equal")
xy_ax.set_xlim(0, 3 * np.pi);
```

```python
# Matplotlib reference
mpl_ax.xaxis.set_major_locator(mpl_plt.MultipleLocator(np.pi / 2))
mpl_ax.xaxis.set_minor_locator(mpl_plt.MultipleLocator(np.pi / 4))
mpl_fig
```

```python
# xy.pyplot
xy_ax.xaxis.set_major_locator(xy_plt.MultipleLocator(np.pi / 2))
xy_ax.xaxis.set_minor_locator(xy_plt.MultipleLocator(np.pi / 4))
xy_fig
```

```python
# Matplotlib reference
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return rf"${N}\pi/2$"
    else:
        return rf"${N // 2}\pi$"


mpl_ax.xaxis.set_major_formatter(mpl_plt.FuncFormatter(format_func))
mpl_fig
```

```python
# xy.pyplot
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return rf"${N}\pi/2$"
    else:
        return rf"${N // 2}\pi$"


xy_ax.xaxis.set_major_formatter(xy_plt.FuncFormatter(format_func))
xy_fig
```
