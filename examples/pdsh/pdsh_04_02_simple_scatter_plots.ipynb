# Simple Scatter Plots

```python
# Matplotlib reference
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")
import numpy as np
```

```python
# xy.pyplot
import xy.pyplot as xy_plt

xy_plt.style.use("default")
import numpy as np
```

## Scatter Plots with plt.plot

```python
# Matplotlib reference
x = np.linspace(0, 10, 30)
y = np.sin(x)

mpl_plt.plot(x, y, "o", color="black");
```

```python
# xy.pyplot
x = np.linspace(0, 10, 30)
y = np.sin(x)

xy_plt.plot(x, y, "o", color="black");
```

```python
# Matplotlib reference
rng = np.random.default_rng(0)
for marker in ["o", ".", ",", "x", "+", "v", "^", "<", ">", "s", "d"]:
    mpl_plt.plot(
        rng.random(2), rng.random(2), marker, color="black", label="marker='{0}'".format(marker)
    )
mpl_plt.legend(numpoints=1, fontsize=13)
mpl_plt.xlim(0, 1.8);
```

```python
# xy.pyplot
rng = np.random.default_rng(0)
for marker in ["o", ".", ",", "x", "+", "v", "^", "<", ">", "s", "d"]:
    xy_plt.plot(
        rng.random(2), rng.random(2), marker, color="black", label="marker='{0}'".format(marker)
    )
xy_plt.legend(numpoints=1, fontsize=13)
xy_plt.xlim(0, 1.8);
```

```python
# Matplotlib reference
mpl_plt.plot(x, y, "-ok");
```

```python
# xy.pyplot
xy_plt.plot(x, y, "-ok");
```

```python
# Matplotlib reference
mpl_plt.plot(
    x,
    y,
    "-p",
    color="gray",
    markersize=15,
    linewidth=4,
    markerfacecolor="white",
    markeredgecolor="gray",
    markeredgewidth=2,
)
mpl_plt.ylim(-1.2, 1.2);
```

```python
# xy.pyplot
xy_plt.plot(
    x,
    y,
    "-p",
    color="gray",
    markersize=15,
    linewidth=4,
    markerfacecolor="white",
    markeredgecolor="gray",
    markeredgewidth=2,
)
xy_plt.ylim(-1.2, 1.2);
```

## Scatter Plots with plt.scatter

```python
# Matplotlib reference
mpl_plt.scatter(x, y, marker="o");
```

```python
# xy.pyplot
xy_plt.scatter(x, y, marker="o");
```

```python
# Matplotlib reference
rng = np.random.default_rng(0)
x = rng.normal(size=100)
y = rng.normal(size=100)
colors = rng.random(100)
sizes = 1000 * rng.random(100)

mpl_plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
mpl_plt.colorbar();  # show color scale
```

```python
# xy.pyplot
rng = np.random.default_rng(0)
x = rng.normal(size=100)
y = rng.normal(size=100)
colors = rng.random(100)
sizes = 1000 * rng.random(100)

xy_plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
xy_plt.colorbar();  # show color scale
```

```python
# Matplotlib reference
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T

mpl_plt.scatter(
    features[0], features[1], alpha=0.4, s=100 * features[3], c=iris.target, cmap="viridis"
)
mpl_plt.xlabel(iris.feature_names[0])
mpl_plt.ylabel(iris.feature_names[1]);
```

```python
# xy.pyplot
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T

xy_plt.scatter(
    features[0], features[1], alpha=0.4, s=100 * features[3], c=iris.target, cmap="viridis"
)
xy_plt.xlabel(iris.feature_names[0])
xy_plt.ylabel(iris.feature_names[1]);
```
