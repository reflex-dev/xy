## Three-Dimensional Points and Lines

```python
# Matplotlib reference
ax = mpl_plt.axes(projection="3d")

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, "gray")

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap="Greens");
```

```python
# xy.pyplot
ax = xy_plt.axes(projection="3d")

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, "gray")

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap="Greens");
```

## Three-Dimensional Contour Plots

```python
# Matplotlib reference
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# xy.pyplot
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# Matplotlib reference
fig = mpl_plt.figure()
ax = mpl_plt.axes(projection="3d")
ax.contour3D(X, Y, Z, 40, cmap="binary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z");
```

```python
# xy.pyplot
fig = xy_plt.figure()
ax = xy_plt.axes(projection="3d")
ax.contour3D(X, Y, Z, 40, cmap="binary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z");
```

```python
# Matplotlib reference
ax.view_init(60, 35)
fig
```

```python
# xy.pyplot
ax.view_init(60, 35)
fig
```

## Wireframes and Surface Plots

```python
# Matplotlib reference
fig = mpl_plt.figure()
ax = mpl_plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z)
ax.set_title("wireframe");
```

```python
# xy.pyplot
fig = xy_plt.figure()
ax = xy_plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z)
ax.set_title("wireframe");
```

```python
# xy.pyplot
ax = xy_plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="viridis", edgecolor="none")
ax.set_title("surface");
```

```python
# Matplotlib reference
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = f(X, Y)

ax = mpl_plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="viridis", edgecolor="none");
```

```python
# xy.pyplot
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = f(X, Y)

ax = xy_plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="viridis", edgecolor="none");
```

## Example: Visualizing a Möbius Strip

```python
# Matplotlib reference
theta = np.linspace(0, 2 * np.pi, 30)
w = np.linspace(-0.25, 0.25, 8)
w, theta = np.meshgrid(w, theta)
```

```python
# xy.pyplot
theta = np.linspace(0, 2 * np.pi, 30)
w = np.linspace(-0.25, 0.25, 8)
w, theta = np.meshgrid(w, theta)
```

```python
# Matplotlib reference
phi = 0.5 * theta
```

```python
# xy.pyplot
phi = 0.5 * theta
```

```python
# Matplotlib reference
# radius in x-y plane
r = 1 + w * np.cos(phi)

x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z = np.ravel(w * np.sin(phi))
```

```python
# xy.pyplot
# radius in x-y plane
r = 1 + w * np.cos(phi)

x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z = np.ravel(w * np.sin(phi))
```

```python
# Matplotlib reference
# triangulate in the underlying parametrization
from matplotlib.tri import Triangulation as mpl_Triangulation  # was matplotlib.tri

tri = mpl_Triangulation(np.ravel(w), np.ravel(theta))

ax = mpl_plt.axes(projection="3d")
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap="Greys", linewidths=0.2)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.axis("off");
```

```python
# xy.pyplot
# triangulate in the underlying parametrization
from xy.pyplot import Triangulation as xy_Triangulation  # was matplotlib.tri

tri = xy_Triangulation(np.ravel(w), np.ravel(theta))

ax = xy_plt.axes(projection="3d")
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap="Greys", linewidths=0.2)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.axis("off");
```
