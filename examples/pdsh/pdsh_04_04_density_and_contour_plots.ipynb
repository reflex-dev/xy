# Density and Contour Plots

```python
# Matplotlib reference
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")
import numpy as np
```

```python
# xy.pyplot
import xy.pyplot as xy_plt

xy_plt.style.use("default")
import numpy as np
```

## Visualizing a Three-Dimensional Function

```python
# Matplotlib reference
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
# xy.pyplot
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
# Matplotlib reference
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# xy.pyplot
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# Matplotlib reference
mpl_plt.contour(X, Y, Z, colors="black");
```

```python
# xy.pyplot
xy_plt.contour(X, Y, Z, colors="black");
```

```python
# Matplotlib reference
mpl_plt.contour(X, Y, Z, 20, cmap="RdGy");
```

```python
# xy.pyplot
xy_plt.contour(X, Y, Z, 20, cmap="RdGy");
```

```python
# Matplotlib reference
mpl_plt.contourf(X, Y, Z, 20, cmap="RdGy")
mpl_plt.colorbar();
```

```python
# xy.pyplot
xy_plt.contourf(X, Y, Z, 20, cmap="RdGy")
xy_plt.colorbar();
```

```python
# Matplotlib reference
mpl_plt.imshow(
    Z, extent=[0, 5, 0, 5], origin="lower", cmap="RdGy", interpolation="gaussian", aspect="equal"
)
mpl_plt.colorbar();
```

```python
# xy.pyplot
xy_plt.imshow(
    Z, extent=[0, 5, 0, 5], origin="lower", cmap="RdGy", interpolation="gaussian", aspect="equal"
)
xy_plt.colorbar();
```

```python
# Matplotlib reference
contours = mpl_plt.contour(X, Y, Z, 3, colors="black")
mpl_plt.clabel(contours, inline=True, fontsize=8)

mpl_plt.imshow(Z, extent=[0, 5, 0, 5], origin="lower", cmap="RdGy", alpha=0.5)
mpl_plt.colorbar();
```

```python
# xy.pyplot
contours = xy_plt.contour(X, Y, Z, 3, colors="black")
xy_plt.clabel(contours, inline=True, fontsize=8)

xy_plt.imshow(Z, extent=[0, 5, 0, 5], origin="lower", cmap="RdGy", alpha=0.5)
xy_plt.colorbar();
```
