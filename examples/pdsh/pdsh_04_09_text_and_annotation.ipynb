## Example: Effect of Holidays on US Births

```python
# Matplotlib reference
# shell command to download the data:
# !cd data && curl -O \
#   https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv
```

```python
# xy.pyplot
# shell command to download the data:
# !cd data && curl -O \
#   https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv
```

```python
# Matplotlib reference
from datetime import datetime

births = pd.read_csv("data/births.csv")

quartiles = np.percentile(births["births"], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query("(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)")

births["day"] = births["day"].astype(int)

births.index = pd.to_datetime(
    10000 * births.year + 100 * births.month + births.day, format="%Y%m%d"
)
births_by_date = births.pivot_table("births", [births.index.month, births.index.day])
births_by_date.index = [datetime(2012, month, day) for (month, day) in births_by_date.index]
```

```python
# xy.pyplot
from datetime import datetime

births = pd.read_csv("data/births.csv")

quartiles = np.percentile(births["births"], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query("(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)")

births["day"] = births["day"].astype(int)

births.index = pd.to_datetime(
    10000 * births.year + 100 * births.month + births.day, format="%Y%m%d"
)
births_by_date = births.pivot_table("births", [births.index.month, births.index.day])
births_by_date.index = [datetime(2012, month, day) for (month, day) in births_by_date.index]
```

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=mpl_ax);
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=xy_ax);
```

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=mpl_ax)

# Add labels to the plot
style = dict(size=10, color="gray")

mpl_ax.text("2012-1-1", 3950, "New Year's Day", **style)
mpl_ax.text("2012-7-4", 4250, "Independence Day", ha="center", **style)
mpl_ax.text("2012-9-4", 4850, "Labor Day", ha="center", **style)
mpl_ax.text("2012-10-31", 4600, "Halloween", ha="right", **style)
mpl_ax.text("2012-11-25", 4450, "Thanksgiving", ha="center", **style)
mpl_ax.text("2012-12-25", 3850, "Christmas ", ha="right", **style)

# Label the axes
mpl_ax.set(title="USA births by day of year (1969-1988)", ylabel="average daily births")

# Format the x-axis with centered month labels
mpl_ax.xaxis.set_major_locator(mpl_mpl.dates.MonthLocator())
mpl_ax.xaxis.set_minor_locator(mpl_mpl.dates.MonthLocator(bymonthday=15))
mpl_ax.xaxis.set_major_formatter(mpl_plt.NullFormatter())
mpl_ax.xaxis.set_minor_formatter(mpl_mpl.dates.DateFormatter("%h"));
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=xy_ax)

# Add labels to the plot
style = dict(size=10, color="gray")

xy_ax.text("2012-1-1", 3950, "New Year's Day", **style)
xy_ax.text("2012-7-4", 4250, "Independence Day", ha="center", **style)
xy_ax.text("2012-9-4", 4850, "Labor Day", ha="center", **style)
xy_ax.text("2012-10-31", 4600, "Halloween", ha="right", **style)
xy_ax.text("2012-11-25", 4450, "Thanksgiving", ha="center", **style)
xy_ax.text("2012-12-25", 3850, "Christmas ", ha="right", **style)

# Label the axes
xy_ax.set(title="USA births by day of year (1969-1988)", ylabel="average daily births")

# Format the x-axis with centered month labels
xy_ax.xaxis.set_major_locator(xy_mpl.dates.MonthLocator())
xy_ax.xaxis.set_minor_locator(xy_mpl.dates.MonthLocator(bymonthday=15))
xy_ax.xaxis.set_major_formatter(xy_plt.NullFormatter())
xy_ax.xaxis.set_minor_formatter(xy_mpl.dates.DateFormatter("%h"));
```

## Transforms and Text Position

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots(facecolor="lightgray")
mpl_ax.axis([0, 10, 0, 10])

# transform=mpl_ax.transData is the default, but we'll specify it anyway
mpl_ax.text(1, 5, ". Data: (1, 5)", transform=mpl_ax.transData)
mpl_ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=mpl_ax.transAxes)
mpl_ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=mpl_fig.transFigure);
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots(facecolor="lightgray")
xy_ax.axis([0, 10, 0, 10])

# transform=xy_ax.transData is the default, but we'll specify it anyway
xy_ax.text(1, 5, ". Data: (1, 5)", transform=xy_ax.transData)
xy_ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=xy_ax.transAxes)
xy_ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=xy_fig.transFigure);
```

```python
# Matplotlib reference
mpl_ax.set_xlim(0, 2)
mpl_ax.set_ylim(-6, 6)
mpl_fig
```

```python
# xy.pyplot
xy_ax.set_xlim(0, 2)
xy_ax.set_ylim(-6, 6)
xy_fig
```

## Arrows and Annotation

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots()

x = np.linspace(0, 20, 1000)
mpl_ax.plot(x, np.cos(x))
mpl_ax.axis("equal")

mpl_ax.annotate(
    "local maximum", xy=(6.28, 1), xytext=(10, 4), arrowprops=dict(facecolor="black", shrink=0.05)
)

mpl_ax.annotate(
    "local minimum",
    xy=(5 * np.pi, -1),
    xytext=(2, -6),
    arrowprops=dict(arrowstyle="->", connectionstyle="angle3,angleA=0,angleB=-90"),
);
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots()

x = np.linspace(0, 20, 1000)
xy_ax.plot(x, np.cos(x))
xy_ax.axis("equal")

xy_ax.annotate(
    "local maximum", xy=(6.28, 1), xytext=(10, 4), arrowprops=dict(facecolor="black", shrink=0.05)
)

xy_ax.annotate(
    "local minimum",
    xy=(5 * np.pi, -1),
    xytext=(2, -6),
    arrowprops=dict(arrowstyle="->", connectionstyle="angle3,angleA=0,angleB=-90"),
);
```

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=mpl_ax)

# Add labels to the plot
mpl_ax.annotate(
    "New Year's Day",
    xy=("2012-1-1", 4100),
    xycoords="data",
    xytext=(50, -30),
    textcoords="offset points",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-0.2"),
)

mpl_ax.annotate(
    "Independence Day",
    xy=("2012-7-4", 4250),
    xycoords="data",
    bbox=dict(boxstyle="round", fc="none", ec="gray"),
    xytext=(10, -40),
    textcoords="offset points",
    ha="center",
    arrowprops=dict(arrowstyle="->"),
)

mpl_ax.annotate(
    "Labor Day Weekend",
    xy=("2012-9-4", 4850),
    xycoords="data",
    ha="center",
    xytext=(0, -20),
    textcoords="offset points",
)
mpl_ax.annotate(
    "",
    xy=("2012-9-1", 4850),
    xytext=("2012-9-7", 4850),
    xycoords="data",
    textcoords="data",
    arrowprops={
        "arrowstyle": "|-|,widthA=0.2,widthB=0.2",
    },
)

mpl_ax.annotate(
    "Halloween",
    xy=("2012-10-31", 4600),
    xycoords="data",
    xytext=(-80, -40),
    textcoords="offset points",
    arrowprops=dict(
        arrowstyle="fancy", fc="0.6", ec="none", connectionstyle="angle3,angleA=0,angleB=-90"
    ),
)

mpl_ax.annotate(
    "Thanksgiving",
    xy=("2012-11-25", 4500),
    xycoords="data",
    xytext=(-120, -60),
    textcoords="offset points",
    bbox=dict(boxstyle="round4,pad=.5", fc="0.9"),
    arrowprops=dict(arrowstyle="->", connectionstyle="angle,angleA=0,angleB=80,rad=20"),
)


mpl_ax.annotate(
    "Christmas",
    xy=("2012-12-25", 3850),
    xycoords="data",
    xytext=(-30, 0),
    textcoords="offset points",
    size=13,
    ha="right",
    va="center",
    bbox=dict(boxstyle="round", alpha=0.1),
    arrowprops=dict(arrowstyle="wedge,tail_width=0.5", alpha=0.1),
)
# Label the axes
mpl_ax.set(title="USA births by day of year (1969-1988)", ylabel="average daily births")

# Format the x-axis with centered month labels
mpl_ax.xaxis.set_major_locator(mpl_mpl.dates.MonthLocator())
mpl_ax.xaxis.set_minor_locator(mpl_mpl.dates.MonthLocator(bymonthday=15))
mpl_ax.xaxis.set_major_formatter(mpl_plt.NullFormatter())
mpl_ax.xaxis.set_minor_formatter(mpl_mpl.dates.DateFormatter("%h"))
mpl_ax.set_ylim(3600, 5400);
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=xy_ax)

# Add labels to the plot
xy_ax.annotate(
    "New Year's Day",
    xy=("2012-1-1", 4100),
    xycoords="data",
    xytext=(50, -30),
    textcoords="offset points",
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=-0.2"),
)

xy_ax.annotate(
    "Independence Day",
    xy=("2012-7-4", 4250),
    xycoords="data",
    bbox=dict(boxstyle="round", fc="none", ec="gray"),
    xytext=(10, -40),
    textcoords="offset points",
    ha="center",
    arrowprops=dict(arrowstyle="->"),
)

xy_ax.annotate(
    "Labor Day Weekend",
    xy=("2012-9-4", 4850),
    xycoords="data",
    ha="center",
    xytext=(0, -20),
    textcoords="offset points",
)
xy_ax.annotate(
    "",
    xy=("2012-9-1", 4850),
    xytext=("2012-9-7", 4850),
    xycoords="data",
    textcoords="data",
    arrowprops={
        "arrowstyle": "|-|,widthA=0.2,widthB=0.2",
    },
)

xy_ax.annotate(
    "Halloween",
    xy=("2012-10-31", 4600),
    xycoords="data",
    xytext=(-80, -40),
    textcoords="offset points",
    arrowprops=dict(
        arrowstyle="fancy", fc="0.6", ec="none", connectionstyle="angle3,angleA=0,angleB=-90"
    ),
)

xy_ax.annotate(
    "Thanksgiving",
    xy=("2012-11-25", 4500),
    xycoords="data",
    xytext=(-120, -60),
    textcoords="offset points",
    bbox=dict(boxstyle="round4,pad=.5", fc="0.9"),
    arrowprops=dict(arrowstyle="->", connectionstyle="angle,angleA=0,angleB=80,rad=20"),
)


xy_ax.annotate(
    "Christmas",
    xy=("2012-12-25", 3850),
    xycoords="data",
    xytext=(-30, 0),
    textcoords="offset points",
    size=13,
    ha="right",
    va="center",
    bbox=dict(boxstyle="round", alpha=0.1),
    arrowprops=dict(arrowstyle="wedge,tail_width=0.5", alpha=0.1),
)
# Label the axes
xy_ax.set(title="USA births by day of year (1969-1988)", ylabel="average daily births")

# Format the x-axis with centered month labels
xy_ax.xaxis.set_major_locator(xy_mpl.dates.MonthLocator())
xy_ax.xaxis.set_minor_locator(xy_mpl.dates.MonthLocator(bymonthday=15))
xy_ax.xaxis.set_major_formatter(xy_plt.NullFormatter())
xy_ax.xaxis.set_minor_formatter(xy_mpl.dates.DateFormatter("%h"))
xy_ax.set_ylim(3600, 5400);
```
