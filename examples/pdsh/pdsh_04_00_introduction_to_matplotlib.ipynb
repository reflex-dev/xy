```python
# Matplotlib reference
import numpy as np

x = np.linspace(0, 10, 100)

mpl_fig = mpl_plt.figure()
mpl_plt.plot(x, np.sin(x), "-")
mpl_plt.plot(x, np.cos(x), "--");
```

### Two Interfaces for the Price of One

#### MATLAB-style Interface

```python
# Matplotlib reference
mpl_plt.figure()  # create a plot figure

# create the first of two panels and set current axis
mpl_plt.subplot(2, 1, 1)  # (rows, columns, panel number)
mpl_plt.plot(x, np.sin(x))

# create the second panel and set current axis
mpl_plt.subplot(2, 1, 2)
mpl_plt.plot(x, np.cos(x));
```

```python
# xy.pyplot
xy_plt.figure()  # create a plot figure

# create the first of two panels and set current axis
xy_plt.subplot(2, 1, 1)  # (rows, columns, panel number)
xy_plt.plot(x, np.sin(x))

# create the second panel and set current axis
xy_plt.subplot(2, 1, 2)
xy_plt.plot(x, np.cos(x));
```

#### Object-oriented interface

```python
# Matplotlib reference
# First create a mpl_grid of plots
# mpl_ax will be an array of two Axes objects
mpl_fig, mpl_ax = mpl_plt.subplots(2)

# Call plot() method on the appropriate object
mpl_ax[0].plot(x, np.sin(x))
mpl_ax[1].plot(x, np.cos(x));
```

```python
# xy.pyplot
# First create a xy_grid of plots
# xy_ax will be an array of two Axes objects
xy_fig, xy_ax = xy_plt.subplots(2)

# Call plot() method on the appropriate object
xy_ax[0].plot(x, np.sin(x))
xy_ax[1].plot(x, np.cos(x));
```
