# Visualizing Uncertainties

## Basic Errorbars

```python
# Matplotlib reference
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")
import numpy as np
```

```python
# xy.pyplot
import xy.pyplot as xy_plt

xy_plt.style.use("default")
import numpy as np
```

```python
# Matplotlib reference
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

mpl_plt.errorbar(x, y, yerr=dy, fmt=".k");
```

```python
# xy.pyplot
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

xy_plt.errorbar(x, y, yerr=dy, fmt=".k");
```

```python
# Matplotlib reference
mpl_plt.errorbar(x, y, yerr=dy, fmt="o", color="black", ecolor="lightgray", elinewidth=3, capsize=0);
```

```python
# xy.pyplot
xy_plt.errorbar(x, y, yerr=dy, fmt="o", color="black", ecolor="lightgray", elinewidth=3, capsize=0);
```

## Continuous Errors

```python
# Matplotlib reference
from sklearn.gaussian_process import GaussianProcessRegressor

# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Compute the Gaussian process fit
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, dyfit = gp.predict(xfit[:, np.newaxis], return_std=True)
```

```python
# xy.pyplot
from sklearn.gaussian_process import GaussianProcessRegressor

# define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Compute the Gaussian process fit
gp = GaussianProcessRegressor()
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, dyfit = gp.predict(xfit[:, np.newaxis], return_std=True)
```

```python
# Matplotlib reference
# Visualize the result
mpl_plt.plot(xdata, ydata, "or")
mpl_plt.plot(xfit, yfit, "-", color="gray")
mpl_plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color="gray", alpha=0.2)
mpl_plt.xlim(0, 10);
```

```python
# xy.pyplot
# Visualize the result
xy_plt.plot(xdata, ydata, "or")
xy_plt.plot(xfit, yfit, "-", color="gray")
xy_plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color="gray", alpha=0.2)
xy_plt.xlim(0, 10);
```
