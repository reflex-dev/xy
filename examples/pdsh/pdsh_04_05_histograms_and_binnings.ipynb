# Histograms, Binnings, and Density

```python
# Matplotlib reference
import numpy as np
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")

rng = np.random.default_rng(1701)
data = rng.normal(size=1000)
```

```python
# xy.pyplot
import numpy as np
import xy.pyplot as xy_plt

xy_plt.style.use("default")

rng = np.random.default_rng(1701)
data = rng.normal(size=1000)
```

```python
# Matplotlib reference
mpl_plt.hist(data);
```

```python
# xy.pyplot
xy_plt.hist(data);
```

```python
# Matplotlib reference
mpl_plt.hist(
    data,
    bins=30,
    density=True,
    alpha=0.5,
    histtype="stepfilled",
    color="steelblue",
    edgecolor="none",
);
```

```python
# xy.pyplot
xy_plt.hist(
    data,
    bins=30,
    density=True,
    alpha=0.5,
    histtype="stepfilled",
    color="steelblue",
    edgecolor="none",
);
```

```python
# Matplotlib reference
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

kwargs = dict(histtype="stepfilled", alpha=0.3, density=True, bins=40)

mpl_plt.hist(x1, **kwargs)
mpl_plt.hist(x2, **kwargs)
mpl_plt.hist(x3, **kwargs);
```

```python
# xy.pyplot
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

kwargs = dict(histtype="stepfilled", alpha=0.3, density=True, bins=40)

xy_plt.hist(x1, **kwargs)
xy_plt.hist(x2, **kwargs)
xy_plt.hist(x3, **kwargs);
```

```python
# Matplotlib reference
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

```python
# xy.pyplot
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

## Two-Dimensional Histograms and Binnings

```python
# Matplotlib reference
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
```

```python
# xy.pyplot
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
```

### plt.hist2d: Two-dimensional histogram

```python
# Matplotlib reference
mpl_plt.hist2d(x, y, bins=30)
cb = mpl_plt.colorbar()
cb.set_label("counts in bin")
```

```python
# xy.pyplot
xy_plt.hist2d(x, y, bins=30)
cb = xy_plt.colorbar()
cb.set_label("counts in bin")
```

```python
# Matplotlib reference
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
print(counts.shape)
```

```python
# xy.pyplot
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
print(counts.shape)
```

### plt.hexbin: Hexagonal binnings

```python
# Matplotlib reference
mpl_plt.hexbin(x, y, gridsize=30)
cb = mpl_plt.colorbar(label="count in bin")
```

```python
# xy.pyplot
xy_plt.hexbin(x, y, gridsize=30)
cb = xy_plt.colorbar(label="count in bin")
```

### Kernel density estimation

```python
# Matplotlib reference
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
mpl_plt.imshow(Z.reshape(Xgrid.shape), origin="lower", aspect="auto", extent=[-3.5, 3.5, -6, 6])
cb = mpl_plt.colorbar()
cb.set_label("density")
```

```python
# xy.pyplot
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
xy_plt.imshow(Z.reshape(Xgrid.shape), origin="lower", aspect="auto", extent=[-3.5, 3.5, -6, 6])
cb = xy_plt.colorbar()
cb.set_label("density")
```
