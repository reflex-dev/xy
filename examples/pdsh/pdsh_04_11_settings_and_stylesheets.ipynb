# Customizing Matplotlib: Configurations and Stylesheets

## Plot Customization by Hand

```python
# Matplotlib reference
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")
import numpy as np
```

```python
# xy.pyplot
import xy.pyplot as xy_plt

xy_plt.style.use("default")
import numpy as np
```

```python
# Matplotlib reference
x = np.random.randn(1000)
mpl_plt.hist(x);
```

```python
# xy.pyplot
x = np.random.randn(1000)
xy_plt.hist(x);
```

```python
# Matplotlib reference
# use a gray background
fig = mpl_plt.figure(facecolor="white")
ax = mpl_plt.axes(facecolor="#E6E6E6")
ax.set_axisbelow(True)

# draw solid white gridlines
mpl_plt.grid(color="w", linestyle="solid")

# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)

# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# lighten ticks and labels
ax.tick_params(colors="gray", direction="out")
for tick in ax.get_xticklabels():
    tick.set_color("gray")
for tick in ax.get_yticklabels():
    tick.set_color("gray")

# control face and edge color of histogram
ax.hist(x, edgecolor="#E6E6E6", color="#EE6666");
```

```python
# xy.pyplot
# use a gray background
fig = xy_plt.figure(facecolor="white")
ax = xy_plt.axes(facecolor="#E6E6E6")
ax.set_axisbelow(True)

# draw solid white gridlines
xy_plt.grid(color="w", linestyle="solid")

# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)

# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# lighten ticks and labels
ax.tick_params(colors="gray", direction="out")
for tick in ax.get_xticklabels():
    tick.set_color("gray")
for tick in ax.get_yticklabels():
    tick.set_color("gray")

# control face and edge color of histogram
ax.hist(x, edgecolor="#E6E6E6", color="#EE6666");
```

## Changing the Defaults: rcParams

```python
# Matplotlib reference
from matplotlib import cycler as mpl_cycler

colors = mpl_cycler("color", ["#EE6666", "#3388BB", "#9988DD", "#EECC55", "#88BB44", "#FFBBBB"])
mpl_plt.rc("figure", facecolor="white")
mpl_plt.rc(
    "axes", facecolor="#E6E6E6", edgecolor="none", axisbelow=True, grid=True, prop_cycle=colors
)
mpl_plt.rc("grid", color="w", linestyle="solid")
mpl_plt.rc("xtick", direction="out", color="gray")
mpl_plt.rc("ytick", direction="out", color="gray")
mpl_plt.rc("patch", edgecolor="#E6E6E6")
mpl_plt.rc("lines", linewidth=2)
```

```python
# xy.pyplot
from xy.pyplot import cycler as xy_cycler

colors = xy_cycler("color", ["#EE6666", "#3388BB", "#9988DD", "#EECC55", "#88BB44", "#FFBBBB"])
xy_plt.rc("figure", facecolor="white")
xy_plt.rc(
    "axes", facecolor="#E6E6E6", edgecolor="none", axisbelow=True, grid=True, prop_cycle=colors
)
xy_plt.rc("grid", color="w", linestyle="solid")
xy_plt.rc("xtick", direction="out", color="gray")
xy_plt.rc("ytick", direction="out", color="gray")
xy_plt.rc("patch", edgecolor="#E6E6E6")
xy_plt.rc("lines", linewidth=2)
```

```python
# Matplotlib reference
mpl_plt.hist(x);
```

```python
# xy.pyplot
xy_plt.hist(x);
```

```python
# Matplotlib reference
for i in range(4):
    mpl_plt.plot(np.random.rand(10))
```

```python
# xy.pyplot
for i in range(4):
    xy_plt.plot(np.random.rand(10))
```

```python
# Matplotlib reference
def hist_and_lines():
    np.random.seed(0)
    fig, ax = mpl_plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(["a", "b", "c"], loc="lower left")
```

```python
# xy.pyplot
def hist_and_lines():
    np.random.seed(0)
    fig, ax = xy_plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(["a", "b", "c"], loc="lower left")
```
