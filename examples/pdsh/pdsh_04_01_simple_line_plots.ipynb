# Simple Line Plots

```python
# Matplotlib reference
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")
import numpy as np
```

```python
# xy.pyplot
import xy.pyplot as xy_plt

xy_plt.style.use("default")
import numpy as np
```

```python
# Matplotlib reference
fig = mpl_plt.figure()
ax = mpl_plt.axes()
```

```python
# xy.pyplot
fig = xy_plt.figure()
ax = xy_plt.axes()
```

```python
# Matplotlib reference
fig = mpl_plt.figure()
ax = mpl_plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
# xy.pyplot
fig = xy_plt.figure()
ax = xy_plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x));
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x));
```

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x))
mpl_plt.plot(x, np.cos(x));
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x))
xy_plt.plot(x, np.cos(x));
```

## Adjusting the Plot: Line Colors and Styles

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x - 0), color="blue")  # specify color by name
mpl_plt.plot(x, np.sin(x - 1), color="g")  # short color code (rgbcmyk)
mpl_plt.plot(x, np.sin(x - 2), color="0.75")  # grayscale between 0 and 1
mpl_plt.plot(x, np.sin(x - 3), color="#FFDD44")  # hex code (RRGGBB, 00 to FF)
mpl_plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  # RGB tuple, values 0 to 1
mpl_plt.plot(x, np.sin(x - 5), color="chartreuse");  # HTML color names supported
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x - 0), color="blue")  # specify color by name
xy_plt.plot(x, np.sin(x - 1), color="g")  # short color code (rgbcmyk)
xy_plt.plot(x, np.sin(x - 2), color="0.75")  # grayscale between 0 and 1
xy_plt.plot(x, np.sin(x - 3), color="#FFDD44")  # hex code (RRGGBB, 00 to FF)
xy_plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  # RGB tuple, values 0 to 1
xy_plt.plot(x, np.sin(x - 5), color="chartreuse");  # HTML color names supported
```

```python
# Matplotlib reference
mpl_plt.plot(x, x + 0, linestyle="solid")
mpl_plt.plot(x, x + 1, linestyle="dashed")
mpl_plt.plot(x, x + 2, linestyle="dashdot")
mpl_plt.plot(x, x + 3, linestyle="dotted")
# For short, you can use the following codes:
mpl_plt.plot(x, x + 4, linestyle="-")  # solid
mpl_plt.plot(x, x + 5, linestyle="--")  # dashed
mpl_plt.plot(x, x + 6, linestyle="-.")  # dashdot
mpl_plt.plot(x, x + 7, linestyle=":");  # dotted
```

```python
# xy.pyplot
xy_plt.plot(x, x + 0, linestyle="solid")
xy_plt.plot(x, x + 1, linestyle="dashed")
xy_plt.plot(x, x + 2, linestyle="dashdot")
xy_plt.plot(x, x + 3, linestyle="dotted")
# For short, you can use the following codes:
xy_plt.plot(x, x + 4, linestyle="-")  # solid
xy_plt.plot(x, x + 5, linestyle="--")  # dashed
xy_plt.plot(x, x + 6, linestyle="-.")  # dashdot
xy_plt.plot(x, x + 7, linestyle=":");  # dotted
```

```python
# Matplotlib reference
mpl_plt.plot(x, x + 0, "-g")  # solid green
mpl_plt.plot(x, x + 1, "--c")  # dashed cyan
mpl_plt.plot(x, x + 2, "-.k")  # dashdot black
mpl_plt.plot(x, x + 3, ":r");  # dotted red
```

```python
# xy.pyplot
xy_plt.plot(x, x + 0, "-g")  # solid green
xy_plt.plot(x, x + 1, "--c")  # dashed cyan
xy_plt.plot(x, x + 2, "-.k")  # dashdot black
xy_plt.plot(x, x + 3, ":r");  # dotted red
```

## Adjusting the Plot: Axes Limits

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x))

mpl_plt.xlim(-1, 11)
mpl_plt.ylim(-1.5, 1.5);
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x))

xy_plt.xlim(-1, 11)
xy_plt.ylim(-1.5, 1.5);
```

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x))

mpl_plt.xlim(10, 0)
mpl_plt.ylim(1.2, -1.2);
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x))

xy_plt.xlim(10, 0)
xy_plt.ylim(1.2, -1.2);
```

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x))
mpl_plt.axis("tight");
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x))
xy_plt.axis("tight");
```

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x))
mpl_plt.axis("equal");
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x))
xy_plt.axis("equal");
```

## Labeling Plots

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x))
mpl_plt.title("A Sine Curve")
mpl_plt.xlabel("x")
mpl_plt.ylabel("sin(x)");
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x))
xy_plt.title("A Sine Curve")
xy_plt.xlabel("x")
xy_plt.ylabel("sin(x)");
```

```python
# Matplotlib reference
mpl_plt.plot(x, np.sin(x), "-g", label="sin(x)")
mpl_plt.plot(x, np.cos(x), ":b", label="cos(x)")
mpl_plt.axis("equal")

mpl_plt.legend();
```

```python
# xy.pyplot
xy_plt.plot(x, np.sin(x), "-g", label="sin(x)")
xy_plt.plot(x, np.cos(x), ":b", label="cos(x)")
xy_plt.axis("equal")

xy_plt.legend();
```

## Matplotlib Gotchas

```python
# Matplotlib reference
ax = mpl_plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel="x", ylabel="sin(x)", title="A Simple Plot");
```

```python
# xy.pyplot
ax = xy_plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel="x", ylabel="sin(x)", title="A Simple Plot");
```
