# Multiple Subplots

```python
# Matplotlib reference
import matplotlib.pyplot as mpl_plt

mpl_plt.style.use("default")
import numpy as np
```

```python
# xy.pyplot
import xy.pyplot as xy_plt

xy_plt.style.use("default")
import numpy as np
```

## plt.axes: Subplots by Hand

```python
# Matplotlib reference
ax1 = mpl_plt.axes()  # standard axes
ax2 = mpl_plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
# xy.pyplot
ax1 = xy_plt.axes()  # standard axes
ax2 = xy_plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
# Matplotlib reference
mpl_fig = mpl_plt.figure()
ax1 = mpl_fig.add_axes([0.1, 0.5, 0.8, 0.4], xticklabels=[], ylim=(-1.2, 1.2))
ax2 = mpl_fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

```python
# xy.pyplot
xy_fig = xy_plt.figure()
ax1 = xy_fig.add_axes([0.1, 0.5, 0.8, 0.4], xticklabels=[], ylim=(-1.2, 1.2))
ax2 = xy_fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

## plt.subplot: Simple Grids of Subplots

```python
# Matplotlib reference
for i in range(1, 7):
    mpl_plt.subplot(2, 3, i)
    mpl_plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha="center")
```

```python
# xy.pyplot
for i in range(1, 7):
    xy_plt.subplot(2, 3, i)
    xy_plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha="center")
```

```python
# Matplotlib reference
mpl_fig = mpl_plt.figure()
mpl_fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    mpl_ax = mpl_fig.add_subplot(2, 3, i)
    mpl_ax.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha="center")
```

```python
# xy.pyplot
xy_fig = xy_plt.figure()
xy_fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    xy_ax = xy_fig.add_subplot(2, 3, i)
    xy_ax.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha="center")
```

## plt.subplots: The Whole Grid in One Go

```python
# Matplotlib reference
mpl_fig, mpl_ax = mpl_plt.subplots(2, 3, sharex="col", sharey="row")
```

```python
# xy.pyplot
xy_fig, xy_ax = xy_plt.subplots(2, 3, sharex="col", sharey="row")
```

```python
# Matplotlib reference
# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
    for j in range(3):
        mpl_ax[i, j].text(0.5, 0.5, str((i, j)), fontsize=18, ha="center")
mpl_fig
```

```python
# xy.pyplot
# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
    for j in range(3):
        xy_ax[i, j].text(0.5, 0.5, str((i, j)), fontsize=18, ha="center")
xy_fig
```

## plt.GridSpec: More Complicated Arrangements

```python
# Matplotlib reference
mpl_grid = mpl_plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
# xy.pyplot
xy_grid = xy_plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
# Matplotlib reference
mpl_plt.subplot(mpl_grid[0, 0])
mpl_plt.subplot(mpl_grid[0, 1:])
mpl_plt.subplot(mpl_grid[1, :2])
mpl_plt.subplot(mpl_grid[1, 2]);
```

```python
# xy.pyplot
xy_plt.subplot(xy_grid[0, 0])
xy_plt.subplot(xy_grid[0, 1:])
xy_plt.subplot(xy_grid[1, :2])
xy_plt.subplot(xy_grid[1, 2]);
```

```python
# Matplotlib reference
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
rng = np.random.default_rng(1701)
x, y = rng.multivariate_normal(mean, cov, 3000).T

# Set up the axes with GridSpec
mpl_fig = mpl_plt.figure(figsize=(6, 6))
mpl_grid = mpl_plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = mpl_fig.add_subplot(mpl_grid[:-1, 1:])
y_hist = mpl_fig.add_subplot(mpl_grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = mpl_fig.add_subplot(mpl_grid[-1, 1:], yticklabels=[], sharex=main_ax)

# Scatter points on the main axes
main_ax.plot(x, y, "ok", markersize=3, alpha=0.2)

# Histogram on the attached axes
x_hist.hist(x, 40, histtype="stepfilled", orientation="vertical", color="gray")
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype="stepfilled", orientation="horizontal", color="gray")
y_hist.invert_xaxis()
```

```python
# xy.pyplot
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
rng = np.random.default_rng(1701)
x, y = rng.multivariate_normal(mean, cov, 3000).T

# Set up the axes with GridSpec
xy_fig = xy_plt.figure(figsize=(6, 6))
xy_grid = xy_plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = xy_fig.add_subplot(xy_grid[:-1, 1:])
y_hist = xy_fig.add_subplot(xy_grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = xy_fig.add_subplot(xy_grid[-1, 1:], yticklabels=[], sharex=main_ax)

# Scatter points on the main axes
main_ax.plot(x, y, "ok", markersize=3, alpha=0.2)

# Histogram on the attached axes
x_hist.hist(x, 40, histtype="stepfilled", orientation="vertical", color="gray")
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype="stepfilled", orientation="horizontal", color="gray")
y_hist.invert_xaxis()
```
